Data loading

```python
import pandas as pd
```

```python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
data = pd.read_csv(r'housing.csv', header=None, delimiter=r"\s+", names=column_names)
```

multicollinearity analysis 

```python
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
# Создание DataFrame с нужными фичами
X = data[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']]

# Добавление константы для расчета VIF
X = add_constant(X)

# Расчет VIF
vif_data = pd.DataFrame()
vif_data['feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif_data)
```

VIF = 1: Нет корреляции между данной фичей и другими.
1 < VIF < 5: Умеренная корреляция.
VIF > 5: Высокая корреляция, стоит рассмотреть удаление фичи.

```python
vif_data[vif_data["VIF"] > 4]
```

Data preprocessing

```python
def preprocess_data(df):
    df = df.drop(['NOX', 'RAD', 'TAX'], axis=1)
    return df
data = preprocess_data(data)
```

Data Split

```python
X = data.drop("MEDV", axis = 1)
y = data["MEDV"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_validation, y_train, y_validation = train_test_split(X, y, train_size=0.8, random_state=1234)
```

Params searching

```python
from catboost import CatBoostRegressor
model = CatBoostRegressor(verbose=0)
```

```python
param_grid = {
    'iterations': [500, 1000],
    'learning_rate': [0.01, 0.1, 0.3],
    'depth': [4, 6, 8],
    'l2_leaf_reg': [1, 3, 5, 7],
    'bagging_temperature': [0.5, 1, 2]
}
grid_search_result = model.grid_search(param_grid, X=X_train, y=y_train, cv=3, plot=True)
```

```python
grid_search_result
```

```python
y_pred = model.predict(X_validation)
```

```python
from sklearn.metrics import mean_squared_error,r2_score

mse = mean_squared_error(y_validation, y_pred)
r2 = r2_score(y_validation, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
```

```python
submission = pd.DataFrame({
    'Id': X_validation.index, 
    'MEDV': y_pred  
})

submission.to_csv('submission.csv', index=False)
```
